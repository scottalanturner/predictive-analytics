# 01 - Welcome to Statistical Thinking

**Key Idea:** Statistics helps us make better decisions when things are uncertain.

## Coin Flip Experiment: Small vs Large Sample

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate coin flips
np.random.seed(42)
small_sample = np.random.choice(['Heads', 'Tails'], size=10)
large_sample = np.random.choice(['Heads', 'Tails'], size=1000)

# Count results
small_heads = np.sum(small_sample == 'Heads')
large_heads = np.sum(large_sample == 'Heads')

# Visualize
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# 10 flips
ax1.bar(['Heads', 'Tails'], [small_heads, 10-small_heads], color=['#4CAF50', '#FF5722'])
ax1.set_title(f'10 Flips: {small_heads} Heads, {10-small_heads} Tails', fontsize=14)
ax1.set_ylabel('Count')
ax1.set_ylim(0, 10)

# 1000 flips
ax2.bar(['Heads', 'Tails'], [large_heads, 1000-large_heads], color=['#4CAF50', '#FF5722'])
ax2.set_title(f'1,000 Flips: {large_heads} Heads, {1000-large_heads} Tails', fontsize=14)
ax2.set_ylabel('Count')
ax2.set_ylim(0, 600)

plt.tight_layout()
plt.show()

print(f"\n10 Flips: {small_heads/10*100:.0f}% heads")
print(f"1,000 Flips: {large_heads/1000*100:.1f}% heads")
print("\n→ More data = clearer patterns!")
```

## Real-World Example: Player Performance Over 20 Games

```python
# Simulate player points per game (noisy but with a trend)
np.random.seed(42)
games = np.arange(1, 21)
points = 15 + np.random.normal(0, 3, 20)  # Average ~15 points, some variability

# Plot
plt.figure(figsize=(10, 5))
plt.plot(games, points, marker='o', linestyle='-', color='#2196F3', linewidth=2, markersize=8)
plt.axhline(y=np.mean(points), color='red', linestyle='--', linewidth=2, label=f'Average: {np.mean(points):.1f} pts')
plt.xlabel('Game Number', fontsize=12)
plt.ylabel('Points Scored', fontsize=12)
plt.title('Player Performance: Points Per Game', fontsize=14)
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\nAverage performance: {np.mean(points):.1f} points per game")
print(f"Range: {np.min(points):.1f} to {np.max(points):.1f}")
print("\n→ Statistics helps predict future performance from past patterns!")
```
